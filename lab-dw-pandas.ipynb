### Data Description

- Customer - Customer ID

- ST - State where customers live

- Gender - Gender of the customer

- Education - Background education of customers 

- Customer Lifetime Value - Customer lifetime value(CLV) is the total revenue the client will derive from their entire relationship with a customer. In other words, is the predicted or calculated value of a customer over their entire duration as a policyholder with the insurance company. It is an estimation of the net profit that the insurance company expects to generate from a customer throughout their relationship with the company. Customer Lifetime Value takes into account factors such as the duration of the customer's policy, premium payments, claim history, renewal likelihood, and potential additional services or products the customer may purchase. It helps insurers assess the long-term profitability and value associated with retaining a particular customer.

- Income - Customers income

- Monthly Premium Auto - Amount of money the customer pays on a monthly basis as a premium for their auto insurance coverage. It represents the recurring cost that the insured person must pay to maintain their insurance policy and receive coverage for potential damages, accidents, or other covered events related to their vehicle.

- Number of Open Complaints - Number of complaints the customer opened

- Policy Type - There are three type of policies in car insurance (Corporate Auto, Personal Auto, and Special Auto)

- Vehicle Class - Type of vehicle classes that customers have Two-Door Car, Four-Door Car SUV, Luxury SUV, Sports Car, and Luxury Car

- Total Claim Amount - the sum of all claims made by the customer. It represents the total monetary value of all approved claims for incidents such as accidents, theft, vandalism, or other covered events.

```python
# Identify the number of unique values for each column and determine which columns appear to be categorical. You should also 
# describe the unique values of each categorical column and the range of values for numerical columns, and give your insights.
print("NUMBER OF UNIQUE VALUES FOR EACH COLUMN:")
print()
for c in df.columns:
    print(f'{c}: {df[c].nunique()}')
print()
print("The columns that appear to be categorical are the ones that have a low number of unique values: ST, GENDER, Education, Number of Open Complaints (we know that it's not categorical from the previous part, but if we only had the information of the unique values we would say it appears to be), Policy Type, Vehicle Class.")
print()
print()
print("UNIQUE VALUES AND RANGES")
print()
for col in df.columns:
    col_type = df[col].dtype
    if col_type=='object':
        print(f'{col}: {df[col].unique()}')
    elif col_type=='float64':
        print(f'{col}: range({df[col].min()}, {df[col].max()})')
print()
print("As we can see in the results obtained, in the columns ST and GENDER there are different representations of the same category (for example, for female there is 'F', 'Femal' and 'female'). We should substitute them with the same term for each category.")
print()
print("There are 'nan' values in each column, but with the results obtained in this part (without looking at the data frame) we cannot observe if there are some empty cells in a row or all the cells in some rows are empty.")
```

```python
# Compute summary statistics such as mean, median, mode, standard deviation, and quartiles to understand the central tendency 
# and distribution of the data for numerical columns. You should also provide your conclusions based on these summary statisti

income = [df['Income'].mean(), df['Income'].median(), df['Income'].mode(), df['Income'].std(), df['Income'].quantile(0.25), df['Income'].quantile(0.5), df['Income'].quantile(0.75)]
monthly_premium_auto=[df['Monthly Premium Auto'].mean(), df['Monthly Premium Auto'].median(), df['Monthly Premium Auto'].mode(), df['Monthly Premium Auto'].std(), df['Monthly Premium Auto'].quantile(0.25), df['Monthly Premium Auto'].quantile(0.5), df['Monthly Premium Auto'].quantile(0.75)]
total_claim_amount=[df['Total Claim Amount'].mean(), df['Total Claim Amount'].median(), df['Total Claim Amount'].mode(), df['Total Claim Amount'].std(), df['Total Claim Amount'].quantile(0.25), df['Total Claim Amount'].quantile(0.5), df['Total Claim Amount'].quantile(0.75)]
df2 = pd.DataFrame(list(zip(income, monthly_premium_auto, total_claim_amount)), index=['mean', 'median', 'mode', 'std', 'quartile1','quartile2','quartile3'], columns =['income', 'monthly_premium_auto', 'total_claim_amount'])
display(df2)
print()
print("We know that the mean is sensitive to extreme values, so we should check if they are representative of each column. To do so, we compare it to the median, which is not sensitive to extreme values.The averages are higher than the medians, especially in the monthly_premium_auto column. In this case, the mean is not only bigger than the median, but is also way bigger than the quartile 3, which is the value under which 75% of the data is found when arranged in increasing order. This means that there are outliers, which have a great impact in the statistical analysis. In the other two columns we can't affirm with certainty if there are extreme values, we should compute the maximum value.")
```

*Note: When analyzing data, we often want to focus on certain groups of values to gain insights. Percentiles are a useful tool to help us define these groups. A percentile is a measure that tells us what percentage of values in a dataset are below a certain value. For example, the 75th percentile represents the value below which 75% of the data falls. Similarly, the 25th percentile represents the value below which 25% of the data falls. When we talk about the top 25%, we are referring to the values that fall above the 75th percentile, which represent the top quarter of the data. On the other hand, when we talk about the bottom 25%, we are referring to the values that fall below the 25th percentile, which represent the bottom quarter of the data. By focusing on these groups, we can identify patterns and trends that may be useful for making decisions and taking action.*

- Use *loc* to create two dataframes: one containing only Personal Auto policies and one containing only Corporate Auto policies.
